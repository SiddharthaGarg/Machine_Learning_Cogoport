```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
```

```python
dataset = pd.read_csv('heart_disease_data.csv')
```

```python
X = dataset.drop(columns='target')
Y = dataset['target']
print(X)
print(Y)
```

```python
scaler = preprocessing.StandardScaler()
X = pd.DataFrame(scaler.fit_transform(X))
print(X)
```

```python
X_train, X_test, Y_train,Y_test = train_test_split(X,Y,train_size=0.8,random_state=2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

models = {
    LogisticRegression(max_iter=100):                    "Logistic Regression",
    SVC():                                   "Support Vector Machine",
    DecisionTreeClassifier(random_state=42): "Decision Tree",
    RandomForestClassifier(n_estimators=10, random_state=42): "Random Forest Classifier"

}

for model in models.keys():
    model.fit(X_train, Y_train)
```

```python
for model, name in models.items():
    print(name + " accuracy on training data" +": {:.2f}%".format(model.score(X_train, Y_train) * 100))
    print(name + " accuracy on test data"+": {:.2f}%".format(model.score(X_test, Y_test) * 100))
```

```python
NN_classifier = MLPClassifier(max_iter=1000)
NN_classifier.fit(X_train, Y_train)
accuracy = NN_classifier.score(X_test, Y_test)
print("Neural Network Accuracy: " + str(accuracy * 100) + "%")
```
