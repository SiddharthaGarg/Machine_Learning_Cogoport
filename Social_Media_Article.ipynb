**PreProcessing**

```python
data.isna()
```

```python
new_data = data.drop(labels=['url',' timedelta'],axis=1)
# new_data.head(4)
# new_data[' n_non_stop_words'].describe()
# new_data[' n_non_stop_words'].median()
new_data = new_data[new_data[' n_tokens_content']!=0]
new_data.info()
# value_counts = new_data[' n_non_stop_words'].value_counts()
# print(value_counts)
```

```python
new_data = new_data.drop(labels=[' n_non_stop_words'],axis=1)
new_data.info()
```

```python
new_data[' weekday_is_monday'] = new_data[' weekday_is_monday']+new_data[' weekday_is_tuesday']+new_data[' weekday_is_wednesday']+ new_data[' weekday_is_thursday'] + new_data[' weekday_is_friday']
# print(data2[' weekday_is_monday'])
new_data = new_data.drop(labels=[' weekday_is_tuesday',' weekday_is_wednesday',' weekday_is_thursday',' weekday_is_friday'],axis=1)
new_data.info()
```

**Label the articles into 'Popular' or 'Unpopular' depending on number of shares**

```python
new_data[' shares'].describe()
new_data[' shares'].median()
```

```python
popularity = new_data[' shares']
```

```python
label_encoder = preprocessing.LabelEncoder()
popular_label = pd.Series(label_encoder.fit_transform(popularity>=1400))
```

```python
print(popular_label)
```

```python
X = new_data.iloc[:,:-1].copy()
Y = popular_label
```

**Feature Scaling**

```python
scaler = preprocessing.StandardScaler()
X = pd.DataFrame(scaler.fit_transform(X))
print(X)
```

**Splitting Data into Training and Test Data**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, train_size=0.8, random_state=123)
```

**Using Different Models**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

models = {
    LogisticRegression(max_iter=1000):                    "Logistic Regression",
    SVC():                                   "Support Vector Machine",
    DecisionTreeClassifier(random_state=42): "Decision Tree",
    RandomForestClassifier(n_estimators=100, random_state=42): "Random Forest Classifier"

}

for model in models.keys():
    model.fit(X_train, Y_train)
```

```python
for model, name in models.items():
    print(name + ": {:.2f}%".format(model.score(X_test, Y_test) * 100))
```

```python
NN_classifier = MLPClassifier(max_iter=500)
NN_classifier.fit(X_train, Y_train)
accuracy = NN_classifier.score(X_test, Y_test)
print("Neural Network Accuracy: " + str(accuracy * 100) + "%")
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
nb_classifier = GaussianNB()
nb_classifier.fit(X_train,Y_train)
print("Gaussian NB Accuracy: " + str(nb_classifier.score(X_test,Y_test)*100)+"%")
```
